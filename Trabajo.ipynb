#1. Escribe un programa que reciba un número e indique si es positivo.

```
import java.util.Scanner;

public class NumeroPositivo {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);

        System.out.print("Ingrese un número: ");
        int numero = scanner.nextInt();

        if (numero > 0) {
            System.out.println("El número es positivo");
        } else {
            System.out.println("El número no es positivo");
        }

        scanner.close();
    }
}


```
#2. Escribe un programa que reciba una edad y determine si la persona es mayor de edad.


```
import java.util.Scanner;

public class MayorDeEdad {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);

        System.out.print("Ingrese su edad: ");
        int edad = scanner.nextInt();

        if (edad >= 18) {
            System.out.println("Es mayor de edad");
        } else {
            System.out.println("Es menor de edad");
        }

        scanner.close();
    }
}



```

#3. Escribe un programa que reciba dos números e imprima el mayor.



```
import java.util.Scanner;

public class CompararNumeros {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);

        System.out.print("Ingrese el primer número: ");
        int num1 = scanner.nextInt();

        System.out.print("Ingrese el segundo número: ");
        int num2 = scanner.nextInt();

        if (num1 > num2) {
            System.out.println("El mayor es: " + num1);
        } else if (num2 > num1) {
            System.out.println("El mayor es: " + num2);
        } else {
            System.out.println("Los números son iguales");
        }

        scanner.close();
    }
}


```


#. Escribe un programa que reciba una calificación (0-100) e indique si es Aprobado (&gt;=60) o Reprobado.
```
import java.util.Scanner;

public class Main {
  public static void main(String[] args) {
    Scanner scanner = new Scanner(System.in);

    System.out.print("Ingrese su calificación (0-100): ");
    int calificacion = scanner.nextInt();

    if (calificacion >= 60) {
      System.out.println("Aprobado");
    } else {
      System.out.println("Reprobado");
    }
  }
}

```
